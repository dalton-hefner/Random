```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
```

```python
from matplotlib.patches import Arc, Circle
def create_court():
    # Set-up figure
    center = plt.imread('/Users/dhefner/Desktop/Kaggle Submission Model/2020MarchMadness.png')

    rect = patches.Rectangle((0, -250), 940, 500, linewidth=0.1,
                             edgecolor='r', facecolor='#dfbb85', zorder=0)
    
    
    fig= plt.figure(figsize=(16, 8))
    ax = fig.add_axes([.2, .1, .6, .8], frame_on=False, 
                       xticks=[], yticks=[])
    ax.add_patch(rect)
    
    #Center court
    newax1 = fig.add_axes([.375, 0.38, 0.25, 0.25], zorder = 1)
    newax1.imshow(center)
    newax1.axis('off')
    
    # Draw the borders of the court
    ax.set_xlim(-20, 960)
    ax.vlines([0, 940], -250, 250)
    ax.hlines([-250, 250], 0, 940)
    ax.hlines([-80, -80, 80, 80], [0, 750] * 2, [190, 940] * 2)
    ax.hlines([-60, -60, 60, 60], [0, 750] * 2, [190, 940] * 2)
    ax.vlines([190, 750], -80, 80)
    ax.vlines(470, -250, 250)
    ax.vlines([40, 900], -30, 30)
    ax.add_patch(Arc((190, 0), 120, 120, theta1=-90, theta2=90))
    ax.add_patch(Arc((190, 0), 120, 120, theta1=90, theta2=-90,linestyle = '--'))
    
    ax.add_patch(Arc((750, 0), 120, 120, theta1=-90, theta2=90,linestyle = '--'))
    
    ax.add_patch(Arc((750, 0), 120, 120, theta1=90, theta2=-90))
    ax.hlines([-220, -220, 220, 220], [0,800]*2,[140, 940]*2)
    ax.add_patch(Arc((892.5, 0), 475, 475, theta1=112.5, theta2 = -112.5))
    
    ax.add_patch(Arc((53.5, 0), 475, 475, theta1=-68.5, theta2 = 68.5))
    
    ax.add_patch(Arc((47.5, 0), 15, 15, theta1=0, theta2=360))
    ax.add_patch(Arc((892.5, 0), 15, 15, theta1=0, theta2=360))
    ax.add_patch(Circle((470, 0), 60, facecolor='none', lw=2))

    
```

```python
create_court()
```
